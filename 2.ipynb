```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dims = (1000,2000)
```

```python
a_real = np.linspace(-2,1,dims[1])
a_imag = (np.linspace(-1,1,dims[0]) * 1j)[None,:].T
plane = a_real + a_imag
```

```python
z = np.full((dims), 0+0j, dtype=np.complex64)
for i in range(200):
    z = z**2 + plane
```

```python
out = z<=2000
plt.matshow(out)
```
