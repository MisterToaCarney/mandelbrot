```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10,10]
```

```python
center = (0,0)
span = 1
extent = [center[0] - span, center[0] + span, center[1] - span, center[1] + span]


dims = (1000,1000)
```

```python
a_real = np.linspace(extent[0], extent[1], dims[1])
a_imag = (np.linspace(extent[2], extent[3], dims[0]) * 1j)[None,:].T
c = a_real + a_imag
```

```python
z = np.full((dims), 0+0j, dtype=np.complex128)
out = np.full((dims), 0, dtype=np.uint64)

for iteration in range(100):
    z = z**2 + c
    out += ((out == 0) & (np.absolute(z) > 2)).astype(np.uint8) * iteration
```

```python
plt.xlabel('real')
plt.ylabel('imaginary')
plt.imshow(out, cmap='inferno', extent=extent)
```
